```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

quakes = pd.read_csv("Almaty_seismic_events_2015_2020.csv")
csv_data = pd.read_csv("preprocessed_Алма_данные.csv")
quakes["Date"] = pd.to_datetime(dict(year=quakes["yr"], month=quakes["Mon"], day=quakes["Day"]))
csv_data["Date"] = pd.to_datetime(csv_data["Date"], errors='coerce')
quakes_daily = quakes.groupby("Date").agg({"M": "mean"}).reset_index().rename(columns={"M": "MeanMagnitude"})
merged_df = pd.merge(quakes_daily, csv_data, on="Date", how="left").dropna()
if merged_df.empty:
    print("Немного пусто")
    merged_df = pd.merge(quakes_daily, csv_data, on="Date", how="left").fillna(merged_df.mean(numeric_only=True))
results = pd.DataFrame([(f, pearsonr(merged_df[f], merged_df["MeanMagnitude"])[0]) for f in merged_df.drop(columns=["Date", "MeanMagnitude"]).select_dtypes(include=["float64", "int64"]).columns], columns=["Feature", "Correlation"])
plt.figure(figsize=(14, 6))
sns.barplot(data=results, x="Feature", y="Correlation")
plt.xticks(rotation=45)
plt.title("Корреляции с MeanMagnitude")
plt.grid(True)
plt.tight_layout()
plt.show()
```
